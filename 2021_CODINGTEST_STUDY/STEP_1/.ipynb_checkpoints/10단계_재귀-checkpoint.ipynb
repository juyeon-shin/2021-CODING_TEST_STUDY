```python
#출력 초과
N = int(input())

star = ['***','* *','***']

def step(star):
    a = []
    for i in range(1,10):
        if i != 5:
            a.append(star)
        else:
            a.append(' ')
    return a

S = star

if N//3 == 1:
    print(star)
elif N//3 != 1:
    while N > 1:
        S = step(S)
        N = N//3
    print(S)
```
