```python
n = int(input())
graph = [[] for _ in range(n + 1)]
parent = [0 for _ in range(n + 1)]

for _ in range(n - 1):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
print(graph) # 2번째 리스트의 0번째 자리가 부모 노드

def bfs():
    from collections import deque 
    q = deque()
    q.append(1)
    while q:
        node = q.popleft()
        for i in graph[node]:
            if parent[i] == 0:
                parent[i] = node
                q.append(i)

    return parent

bfs()

for i in parent[2:]:
    print(i)
```

```python
from sys import stdin
from collections import deque

#read = stdin.readline()
V = int(input())
graph = [[] for _ in range(V + 1)]

for _ in range(V):
    c = list(map(int,input().split()))
    for e in range(1, len(c) - 2, 2):
        graph[c[0]].append((c[e], c[e + 1])) # 그래프에 노드 간선 맞춰서 리스트에 넣기
print(graph)

def bfs(start):
    visit = [-1] * (V + 1)
    que = deque()
    que.append(start)
    visit[start] = 0
    _max = [0, 0]

    while que: # 여기서부터 이해안감
        t = que.popleft()
        for e, w in graph[t]: # graph 안 원소가 []로 되어있어서 가능
            if visit[e] == -1:
                print(que, visit)
                visit[e] = visit[t] + w
                que.append(e)
                print(que,visit)
                if _max[0] < visit[e]:
                    _max = visit[e], e
                    print(_max)

    return _max


dis, node = bfs(1)
dis, node = bfs(node)
print(dis)
```

```python
from collections import deque
n=int(input())
tree={i:[] for i in range(n+1)} # 딕셔너리 처리
for i in range(n-1):
    a,b,weight=map(int, input().split())
    tree[a].append((b,weight))
    tree[b].append((a,weight))
print(tree)

def bfs(s):
    queue=deque()
    queue.append((s,0))
    visited=[0]*n
    visited[s-1]=1

    result=[0,0]
    while queue:
        now,cnt=queue.popleft() # 1,0 -> 2,0
        print(now,cnt)
        for i in tree[now]:
            next_number,value=i[0],i[1] # next_ = 2->1, values = 3-> 3
            if visited[next_number-1]==0:
                visited[next_number-1]=1 # visited[1] = 1
                queue.append((next_number,cnt+value))
            if result[1]<cnt+value: # cnt+value = 2+3=5
                result[0]=next_number # result[0] = 2
                result[1]=cnt+value # result[1] = 0+3
                print(result) # result = [2,3]

        return result

a=bfs(1)
b=bfs(a[0])
print(b[1])
```
